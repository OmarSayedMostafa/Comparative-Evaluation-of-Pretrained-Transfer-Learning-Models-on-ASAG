## Pearson correlation

```
x = [1,2,3,4]
y_1 = [2,3,4,5]
y_2 = [1,4,3,5]
y_3 = [5,4,3,2]
```

```
def pearson_correlation(x,y):
        mean_x = sum(x) / len(x)
        mean_y = sum(y) / len(y)  
        cov = sum((a - mean_x) * (b - mean_y) for (a,b) in zip(x,y)) / len(x)
        
        std_x, std_y = np.std(x), np.std(y)

        p = cov/(std_x*std_y)

        return float(p)
```

```
pc_1 = pearson_correlation(x, y_1)
pc_2 = pearson_correlation(x, y_2)
pc_3 = pearson_correlation(x, y_3)

print(pc_1, pc_2, pc_3)
```

```
fig = plt.figure()
ax = fig.add_subplot(111)

sns.regplot(x, y_1)

ax.grid()
plt.xlabel('y', fontsize=14, weight='bold')
plt.ylabel('Å·', fontsize=14, weight='bold')
plt.title("Pearson correlation = 1.000", fontsize=14, weight='bold')

plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)

sns.regplot(x, y_2, ci= None)

ax.grid(linestyle='--')
plt.xlabel('y', fontsize=14, weight='bold')
plt.ylabel('Å·', fontsize=14, weight='bold')
plt.title("Pearson correlation = 0.832", fontsize=14, weight='bold')
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(111)

sns.regplot(x, y_3)

ax.grid(linestyle='--')
plt.xlabel('y', fontsize=14, weight='bold')
plt.ylabel('Å·', fontsize=14, weight='bold')
plt.title("Pearson correlation = -1.000", fontsize=14, weight='bold')
plt.show()
```

## Spearman coefficient

```
from scipy.stats import spearmanr
```

```
sc_x1 = np.linspace(-1,1,50)
sc_y1 = sc_x1**3

sc_x2 = np.random.random_sample(50)
sc_y2 = np.random.random_sample(50)

print(spearmanr(sc_x1, sc_y1))
print(spearmanr(sc_x2, sc_y2))
```
