```
import wikipediaapi
```

```
wiki_wiki = wikipediaapi.Wikipedia('en', extract_format=wikipediaapi.ExtractFormat.WIKI)
page_py = wiki_wiki.page("computer science")
```

```
print("Page exists- ", page_py.exists())
```

```
def print_links(page):
  links = page.links
  for title in sorted(links.keys()):
    print(title, links[title])
```

```
print(print_links(page_py))
```
